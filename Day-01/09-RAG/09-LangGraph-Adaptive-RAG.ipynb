# Adaptive RAG

이 튜토리얼은 Adaptive RAG(Adaptive Retrieval-Augmented Generation)의 구현을 다룹니다. 

Adaptive RAG는 쿼리 분석과 능동적/자기 수정 RAG를 결합하여 다양한 데이터 소스에서 정보를 검색하고 생성하는 전략입니다. 

이 튜토리얼에서는 LangGraph를 사용하여 웹 검색과 자기 수정 RAG 간의 라우팅을 구현합니다.

**주로 다루는 내용**

- **Create Index**: 인덱스 생성 및 문서 로드
- **LLMs**: LLM을 사용한 쿼리 라우팅 및 문서 평가
- **Web Search Tool**: 웹 검색 도구 설정
- **Construct the Graph**: 그래프 상태 및 흐름 정의
- **Compile Graph**: 그래프 컴파일 및 워크플로우 구축
- **Use Graph**: 그래프 실행 및 결과 확인

----

**Adaptive RAG**는 **RAG**의 전략으로, (1) [쿼리 분석](https://blog.langchain.dev/query-construction/)과 (2) [Self-Reflective RAG](https://blog.langchain.dev/agentic-rag-with-langgraph/)을 결합합니다.

[논문: Adaptive-RAG: Learning to Adapt Retrieval-Augmented Large Language Models through Question Complexity](https://arxiv.org/abs/2403.14403) 에서는 쿼리 분석을 통해 다음과 같은 라우팅을 수행합니다.

- `No Retrieval`
- `Single-shot RAG`
- `Iterative RAG`

LangGraph를 사용하여 이를 구현합니다.

이 구현에서는 다음과 같은 라우팅을 수행합니다.

- **웹 검색**: 최신 이벤트와 관련된 질문에 사용
- **자기 수정 RAG**: 인덱스와 관련된 질문에 사용

## 쿼리 라우팅과 문서 평가

**LLMs** 단계에서는 **쿼리 라우팅**과 **문서 평가**를 수행합니다. 이 과정은 **Adaptive RAG**의 중요한 부분으로, 효율적인 정보 검색과 생성에 기여합니다.

- **쿼리 라우팅**: 사용자의 쿼리를 분석하여 적절한 정보 소스로 라우팅합니다. 이를 통해 쿼리의 목적에 맞는 최적의 검색 경로를 설정할 수 있습니다.
- **문서 평가**: 검색된 문서의 품질과 관련성을 평가하여 최종 결과의 정확성을 높입니다. 이 과정은 **LLMs**의 성능을 극대화하는 데 필수적입니다.

이 단계는 **Adaptive RAG**의 핵심 기능을 지원하며, 정확하고 신뢰할 수 있는 정보 제공을 목표로 합니다.

아래 코드에서는 `RouteQuery` 데이터 모델과 `question_router` 체인을 정의합니다.

```python
from typing import Literal

from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field
from langchain.chat_models import init_chat_model


# 사용자 쿼리를 가장 관련성 높은 데이터 소스로 라우팅하는 데이터 모델
class RouteQuery(BaseModel):
    """사용자 질문을 가장 적합한 데이터 소스(벡터스토어 또는 웹 검색)로 라우팅하는 모델입니다."""

    # 데이터 소스 선택을 위한 리터럴 타입 필드
    datasource: Literal["vectorstore", "web_search"] = Field(
        ...,
        description="Given a user question choose to route it to web search or a vectorstore.",
    )


# LLM 초기화 (OpenAI 키 사용 시 gpt-5.2, gpt-4.1-mini 등의 모델로 변경하세요.)
llm = init_chat_model("claude-sonnet-4-5")
# 구조화된 출력을 위한 라우터 LLM 설정
structured_llm_router = llm.with_structured_output(RouteQuery)

# 시스템 메시지와 사용자 질문을 포함한 프롬프트 템플릿 생성
system = """You are an expert at routing a user question to a vectorstore or web search.
The vectorstore contains documents related to DEC 2023 AI Brief Report(SPRI) with Samsung Gauss, Anthropic, etc.
Use the vectorstore for questions on these topics. Otherwise, use web-search."""

# Routing 을 위한 프롬프트 템플릿 생성
route_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system),
        ("human", "{question}"),
    ]
)

# 프롬프트 템플릿과 구조화된 LLM 라우터를 결합하여 질문 라우터 생성
question_router = route_prompt | structured_llm_router
```

다음은 쿼리 라우팅 결과를 테스트 해본 뒤 결과를 확인합니다.

```python
# 문서 검색이 필요한 질문
print(
    question_router.invoke(
        {"question": "AI Brief 에서 삼성전자가 만든 생성형 AI 의 이름은?"}
    )
)
```

```python
# 웹 검색이 필요한 질문
print(question_router.invoke({"question": "판교에서 가장 맛있는 딤섬집 찾아줘"}))
```

### 검색 평가기(Retrieval Grader)

**검색 평가기(Retrieval Grader)**는 벡터스토어에서 검색된 문서가 사용자 질문과 실제로 관련이 있는지 평가하는 구성 요소입니다. 벡터스토어 검색은 의미적 유사도 기반이므로 관련 없는 문서가 포함될 수 있으며, 이 평가기를 통해 불필요한 문서를 필터링하여 답변 품질을 향상시킵니다.

평가 결과는 `yes` 또는 `no`의 이진 점수로 반환됩니다. `yes`는 문서가 질문과 관련 있음을, `no`는 관련 없음을 의미합니다. 관련 없는 문서가 모두 필터링되면 쿼리를 재작성하여 검색을 다시 시도합니다.

아래 코드에서는 `GradeDocuments` 데이터 모델과 `retrieval_grader` 체인을 정의합니다.

```python
from pydantic import BaseModel, Field
from langchain_core.prompts import ChatPromptTemplate
from langchain.chat_models import init_chat_model


# 문서 평가를 위한 데이터 모델 정의
class GradeDocuments(BaseModel):
    """검색된 문서가 사용자 질문과 관련이 있는지 평가하는 이진 점수 모델입니다."""

    binary_score: str = Field(
        description="Documents are relevant to the question, 'yes' or 'no'"
    )


# LLM 초기화 (OpenAI 키 사용 시 gpt-5.2, gpt-4.1-mini 등의 모델로 변경하세요.)
llm = init_chat_model("claude-sonnet-4-5")
# 구조화된 출력을 위한 평가기 LLM 설정
structured_llm_grader = llm.with_structured_output(GradeDocuments)

# 시스템 메시지와 사용자 질문을 포함한 프롬프트 템플릿 생성
system = """You are a grader assessing relevance of a retrieved document to a user question. \n 
    If the document contains keyword(s) or semantic meaning related to the user question, grade it as relevant. \n
    It does not need to be a stringent test. The goal is to filter out erroneous retrievals. \n
    Give a binary score 'yes' or 'no' score to indicate whether the document is relevant to the question."""

grade_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system),
        ("human", "Retrieved document: \n\n {document} \n\n User question: {question}"),
    ]
)

# 문서 검색결과 평가기 생성
retrieval_grader = grade_prompt | structured_llm_grader
```

생성한 `retrieval_grader` 를 사용하여 문서 검색결과를 평가합니다.

```python
# 사용자 질문 설정
question = "삼성전자가 만든 생성형 AI 의 이름은?"

# 질문에 대한 관련 문서 검색
docs = pdf_retriever.invoke(question)
```

```python
# 검색된 문서의 내용 가져오기
retrieved_doc = docs[1].page_content

# 평가 결과 출력
print(retrieval_grader.invoke({"question": question, "document": retrieved_doc}))
```

```python
# 필터링 하는 코드 예시
filtered_docs = []
for doc in docs:
    result = retrieval_grader.invoke(
        {
            "question": question,
            "document": doc.page_content,
        }
    )
    if result.binary_score == "yes":
        filtered_docs.append(doc)
```

### 답변 생성을 위한 RAG 체인 생성

**RAG 체인(RAG Chain)**은 검색된 문서를 컨텍스트로 활용하여 LLM이 더 정확한 답변을 생성하도록 합니다. 검색된 문서를 `format_docs` 함수로 XML 형식의 문자열로 포맷팅한 후 프롬프트에 전달하면, LLM은 문서 내용을 바탕으로 답변을 생성합니다.

`teddynote/rag-prompt`는 LangChain Hub에서 제공하는 RAG 전용 프롬프트 템플릿으로, 검색된 문서와 질문을 구조화된 형식으로 LLM에 전달합니다. 이 프롬프트는 필요에 따라 자유롭게 수정할 수 있습니다.

아래 코드에서는 Hub에서 프롬프트를 가져오고, `format_docs` 함수와 함께 RAG 체인을 구성합니다.

```python
from langsmith import Client
from langchain_core.output_parsers import StrOutputParser
from langchain.chat_models import init_chat_model

# LangSmith Client를 사용하여 프롬프트 가져오기(RAG 프롬프트는 자유롭게 수정 가능)
prompt = Client().pull_prompt("teddynote/rag-prompt")

# LLM 초기화 (OpenAI 키 사용 시 gpt-5.2, gpt-4.1-mini 등의 모델로 변경하세요.)
llm = init_chat_model("claude-sonnet-4-5")


def format_docs(docs):
    """검색된 문서 목록을 XML 형식의 문자열로 포맷팅합니다.

    각 문서의 내용, 출처, 페이지 번호를 포함한 구조화된 형식으로 변환하여
    LLM이 문서 내용을 명확히 구분하고 참조할 수 있도록 합니다.
    """
    return "\n\n".join(
        [
            f'<document><content>{doc.page_content}</content><source>{doc.metadata["source"]}</source><page>{doc.metadata["page"]+1}</page></document>'
            for doc in docs
        ]
    )


# RAG 체인 생성
rag_chain = prompt | llm | StrOutputParser()
```

이제 생성한 `rag_chain` 에 질문을 전달하여 답변을 생성합니다.

```python
# RAG 체인에 질문을 전달하여 답변 생성
generation = rag_chain.invoke({"context": format_docs(docs), "question": question})
print(generation)
```

### 답변의 Hallucination 체커 추가

**할루시네이션(Hallucination)**은 LLM이 검색된 문서에 없는 내용을 사실인 것처럼 생성하는 현상입니다. 이 체커는 생성된 답변이 실제로 검색된 문서(사실)에 근거하고 있는지를 평가하여, 신뢰성 없는 답변이 최종 결과로 반환되는 것을 방지합니다.

평가 결과가 `yes`이면 답변이 문서에 근거한 것으로 판단하고 다음 단계로 진행합니다. `no`이면 환각이 발생한 것으로 간주하여 답변 생성 노드로 돌아가 재시도합니다.

아래 코드에서는 `GradeHallucinations` 데이터 모델과 `hallucination_grader` 체인을 정의합니다.

```python
from pydantic import BaseModel, Field
from langchain_core.prompts import ChatPromptTemplate
from langchain.chat_models import init_chat_model


# 할루시네이션 체크를 위한 데이터 모델 정의
class GradeHallucinations(BaseModel):
    """생성된 답변이 검색된 사실에 근거하고 있는지 평가하는 이진 점수 모델입니다."""

    binary_score: str = Field(
        description="Answer is grounded in the facts, 'yes' or 'no'"
    )


# LLM 초기화 (OpenAI 키 사용 시 gpt-5.2, gpt-4.1-mini 등의 모델로 변경하세요.)
llm = init_chat_model("claude-sonnet-4-5")
# 구조화된 출력을 위한 평가기 LLM 설정
structured_llm_grader = llm.with_structured_output(GradeHallucinations)

# 프롬프트 설정
system = """You are a grader assessing whether an LLM generation is grounded in / supported by a set of retrieved facts. \n 
    Give a binary score 'yes' or 'no'. 'Yes' means that the answer is grounded in / supported by the set of facts."""

# 프롬프트 템플릿 생성
hallucination_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system),
        ("human", "Set of facts: \n\n {documents} \n\n LLM generation: {generation}"),
    ]
)

# 환각 평가기 생성
hallucination_grader = hallucination_prompt | structured_llm_grader
```

생성한 `hallucination_grader` 를 사용하여 생성된 답변의 환각 여부를 평가합니다.

```python
# 평가기를 사용하여 생성된 답변의 환각 여부 평가
hallucination_grader.invoke({"documents": docs, "generation": generation})
```

### 답변의 관련성 평가기(Answer Grader)

**답변 관련성 평가기(Answer Grader)**는 생성된 답변이 사용자의 질문을 실제로 해결하는지 최종적으로 평가합니다. 할루시네이션 체크를 통과한 답변이라도 질문의 의도와 맞지 않는 경우가 있을 수 있으며, 이 평가기는 그러한 경우를 걸러냅니다.

평가 결과가 `yes`이면 질문이 해결된 것으로 최종 답변을 반환하고, `no`이면 쿼리를 재작성하여 검색부터 다시 시도합니다.

아래 코드에서는 `GradeAnswer` 데이터 모델과 `answer_grader` 체인을 정의합니다.

```python
from pydantic import BaseModel, Field
from langchain_core.prompts import ChatPromptTemplate
from langchain.chat_models import init_chat_model


class GradeAnswer(BaseModel):
    """생성된 답변이 사용자 질문을 실제로 해결하는지 평가하는 이진 점수 모델입니다."""

    binary_score: str = Field(
        description="Indicate 'yes' or 'no' whether the answer solves the question"
    )


# LLM 초기화 (OpenAI 키 사용 시 gpt-5.2, gpt-4.1-mini 등의 모델로 변경하세요.)
llm = init_chat_model("claude-sonnet-4-5")
# 구조화된 출력을 위한 평가기 LLM 설정
structured_llm_grader = llm.with_structured_output(GradeAnswer)

# 프롬프트 설정
system = """You are a grader assessing whether an answer addresses / resolves a question \n 
     Give a binary score 'yes' or 'no'. Yes' means that the answer resolves the question."""
answer_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system),
        ("human", "User question: \n\n {question} \n\n LLM generation: {generation}"),
    ]
)

# 프롬프트 템플릿과 구조화된 LLM 평가기를 결합하여 답변 평가기 생성
answer_grader = answer_prompt | structured_llm_grader
```

```python
# 평가기를 사용하여 생성된 답변이 질문을 해결하는지 여부 평가
answer_grader.invoke({"question": question, "generation": generation})
```

### 쿼리 재작성(Query Rewriter)

**쿼리 재작성기(Query Rewriter)**는 검색 품질을 향상시키기 위해 사용자의 원래 질문을 벡터스토어 검색에 더 적합한 형태로 변환합니다. 문서에서 관련 내용을 찾지 못하거나 생성된 답변이 질문에 맞지 않는 경우, 재작성된 질문으로 검색을 다시 시도합니다.

예를 들어, "삼성전자 AI 이름은?" 같은 간략한 질문을 "삼성전자가 개발한 생성형 AI의 모델명"처럼 검색 최적화된 형태로 확장하여 더 관련성 높은 문서를 검색할 수 있습니다.

아래 코드에서는 질문 재작성을 위한 `question_rewriter` 체인을 생성합니다.

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain.chat_models import init_chat_model

# LLM 초기화 (OpenAI 키 사용 시 gpt-5.2, gpt-4.1-mini 등의 모델로 변경하세요.)
llm = init_chat_model("claude-sonnet-4-5")

# Query Rewriter 프롬프트 정의(자유롭게 수정이 가능합니다)
system = """You a question re-writer that converts an input question to a better version that is optimized \n 
for vectorstore retrieval. Look at the input and try to reason about the underlying semantic intent / meaning."""

# Query Rewriter 프롬프트 템플릿 생성
re_write_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system),
        (
            "human",
            "Here is the initial question: \n\n {question} \n Formulate an improved question.",
        ),
    ]
)

# Query Rewriter 생성
question_rewriter = re_write_prompt | llm | StrOutputParser()
```

생성한 `question_rewriter` 에 질문을 전달하여 개선된 질문을 생성합니다.

```python
# 질문 재작성기에 질문을 전달하여 개선된 질문 생성
question_rewriter.invoke({"question": question})
```

## 그래프 구성

지금까지 정의한 컴포넌트들(라우터, 평가기, RAG 체인, 웹 검색 도구)을 LangGraph의 `StateGraph`로 통합합니다. 그래프는 상태(State), 노드(Node), 엣지(Edge)로 구성되며, 각 노드가 상태를 읽고 업데이트하면서 워크플로우가 진행됩니다.

> 📖 **참고 문서**: [LangGraph Graph API](https://docs.langchain.com/oss/python/langgraph/graph-api.md)

### 그래프 상태 정의

**GraphState**는 LangGraph 워크플로우 전반에 걸쳐 공유되는 데이터 구조입니다. 각 노드는 이 상태를 읽어 처리하고, 업데이트된 값을 딕셔너리 형태로 반환합니다. LangGraph v1에서는 `TypedDict` 기반의 상태 정의만 지원합니다.

Adaptive RAG의 상태는 세 가지 핵심 필드로 구성됩니다. `question`은 사용자 질문, `generation`은 LLM이 생성한 답변, `documents`는 검색된 문서 목록입니다. 각 노드가 필요한 필드만 반환하면 나머지 필드는 이전 값을 유지합니다.

아래 코드에서는 `GraphState`를 `TypedDict`로 정의합니다.

```python
from typing import List
from typing_extensions import TypedDict, Annotated


# 그래프의 상태 정의
class GraphState(TypedDict):
    """
    그래프의 상태를 나타내는 데이터 모델

    Attributes:
        question: 질문
        generation: LLM 생성된 답변
        documents: 도큐먼트 리스트
    """

    question: Annotated[str, "User question"]
    generation: Annotated[str, "LLM generated answer"]
    documents: Annotated[List[str], "List of documents"]
```

## 그래프 흐름 정의

**그래프 흐름**을 정의하여 **Adaptive RAG**의 작동 방식을 명확히 합니다. 이 단계에서는 그래프의 상태와 전환을 설정하여 쿼리 처리의 효율성을 높입니다.

- **상태 정의**: 그래프의 각 상태를 명확히 정의하여 쿼리의 진행 상황을 추적합니다.
- **전환 설정**: 상태 간의 전환을 설정하여 쿼리가 적절한 경로를 따라 진행되도록 합니다.
- **흐름 최적화**: 그래프의 흐름을 최적화하여 정보 검색과 생성의 정확성을 향상시킵니다.

### 노드 정의

각 노드(Node)는 `GraphState`를 입력으로 받아 처리한 후, 업데이트할 상태 필드를 딕셔너리로 반환하는 함수입니다. Adaptive RAG에서는 다음 다섯 가지 노드를 정의합니다.

- **retrieve**: PDF 벡터스토어에서 질문과 관련된 문서를 검색합니다.
- **generate**: 검색된 문서를 컨텍스트로 LLM이 답변을 생성합니다.
- **grade_documents**: 검색된 각 문서의 관련성을 평가하고 관련 없는 문서를 필터링합니다.
- **transform_query**: 검색 품질 향상을 위해 질문을 재작성합니다.
- **web_search**: Tavily API를 사용해 웹에서 최신 정보를 검색합니다.

아래 코드에서는 다섯 가지 노드 함수를 정의합니다.

```python
from langchain_core.documents import Document


# 문서 검색 노드
def retrieve(state):
    """벡터스토어에서 질문과 관련된 문서를 검색하는 노드입니다.

    Args:
        state: question 필드를 포함한 그래프 상태

    Returns:
        documents 필드가 업데이트된 상태 딕셔너리
    """
    print("==== [RETRIEVE] ====")
    question = state["question"]

    # 문서 검색 수행
    documents = pdf_retriever.invoke(question)
    return {"documents": documents}


# 답변 생성 노드
def generate(state):
    """검색된 문서를 컨텍스트로 활용하여 LLM이 답변을 생성하는 노드입니다.

    Args:
        state: question, documents 필드를 포함한 그래프 상태

    Returns:
        generation 필드가 업데이트된 상태 딕셔너리
    """
    print("==== [GENERATE] ====")
    # 질문과 문서 검색 결과 가져오기
    question = state["question"]
    documents = state["documents"]

    # RAG 답변 생성
    generation = rag_chain.invoke({"context": documents, "question": question})
    return {"generation": generation}


# 문서 관련성 평가 노드
def grade_documents(state):
    """검색된 각 문서의 관련성을 평가하고 관련 없는 문서를 필터링하는 노드입니다.

    Args:
        state: question, documents 필드를 포함한 그래프 상태

    Returns:
        관련 문서만 포함된 documents 필드가 업데이트된 상태 딕셔너리
    """
    print("==== [CHECK DOCUMENT RELEVANCE TO QUESTION] ====")
    # 질문과 문서 검색 결과 가져오기
    question = state["question"]
    documents = state["documents"]

    # 각 문서에 대한 관련성 점수 계산
    filtered_docs = []
    for d in documents:
        score = retrieval_grader.invoke(
            {"question": question, "document": d.page_content}
        )
        grade = score.binary_score
        if grade == "yes":
            print("---GRADE: DOCUMENT RELEVANT---")
            # 관련성이 있는 문서 추가
            filtered_docs.append(d)
        else:
            # 관련성이 없는 문서는 건너뛰기
            print("---GRADE: DOCUMENT NOT RELEVANT---")
            continue
    return {"documents": filtered_docs}


# 질문 재작성 노드
def transform_query(state):
    """벡터스토어 검색에 더 적합하도록 질문을 재작성하는 노드입니다.

    Args:
        state: question 필드를 포함한 그래프 상태

    Returns:
        재작성된 질문으로 업데이트된 question 필드를 포함한 상태 딕셔너리
    """
    print("==== [TRANSFORM QUERY] ====")
    # 질문 가져오기
    question = state["question"]

    # 질문 재작성
    better_question = question_rewriter.invoke({"question": question})
    return {"question": better_question}


# 웹 검색 노드
def web_search(state):
    """Tavily API를 사용하여 웹에서 최신 정보를 검색하는 노드입니다.

    Args:
        state: question 필드를 포함한 그래프 상태

    Returns:
        웹 검색 결과로 업데이트된 documents 필드를 포함한 상태 딕셔너리
    """
    print("==== [WEB SEARCH] ====")
    # 질문 가져오기
    question = state["question"]

    # 웹 검색 수행
    web_results = web_search_tool.invoke({"query": question})
    web_results_docs = [
        Document(
            page_content=web_result["content"],
            metadata={"source": web_result["url"]},
        )
        for web_result in web_results
    ]

    return {"documents": web_results_docs}
```

## 조건부 엣지 함수 정의

**조건부 엣지(Conditional Edge)**는 현재 상태를 기반으로 다음에 실행할 노드를 동적으로 결정하는 함수입니다. 일반 엣지와 달리 조건에 따라 서로 다른 노드로 분기할 수 있어 복잡한 워크플로우를 구현할 때 필수적입니다.

Adaptive RAG에서는 세 가지 조건부 엣지 함수를 사용합니다.

- **route_question**: 질문의 성격을 분석하여 웹 검색 또는 벡터스토어 검색으로 라우팅합니다.
- **decide_to_generate**: 관련 문서 존재 여부에 따라 답변 생성 또는 쿼리 재작성을 결정합니다.
- **hallucination_check**: 생성된 답변의 환각 여부와 질문 해결 여부를 순차적으로 평가하여 다음 단계를 결정합니다.

아래 코드에서는 세 가지 조건부 엣지 함수를 정의합니다.

```python
# 질문 라우팅 조건부 엣지 함수
def route_question(state):
    """질문의 성격에 따라 웹 검색 또는 벡터스토어로 라우팅하는 조건부 엣지 함수입니다.

    Args:
        state: question 필드를 포함한 그래프 상태

    Returns:
        다음 노드 이름 ("web_search" 또는 "vectorstore")
    """
    print("==== [ROUTE QUESTION] ====")
    # 질문 가져오기
    question = state["question"]
    # 질문 라우팅
    source = question_router.invoke({"question": question})
    # 질문 라우팅 결과에 따른 노드 라우팅
    if source.datasource == "web_search":
        print("==== [ROUTE QUESTION TO WEB SEARCH] ====")
        return "web_search"
    elif source.datasource == "vectorstore":
        print("==== [ROUTE QUESTION TO VECTORSTORE] ====")
        return "vectorstore"


# 문서 평가 후 생성 여부를 결정하는 조건부 엣지 함수
def decide_to_generate(state):
    """관련 문서 존재 여부에 따라 답변 생성 또는 쿼리 재작성을 결정하는 조건부 엣지 함수입니다.

    Args:
        state: documents 필드를 포함한 그래프 상태

    Returns:
        다음 노드 이름 ("transform_query" 또는 "generate")
    """
    print("==== [DECISION TO GENERATE] ====")
    # 필터링된 문서 검색 결과 가져오기
    filtered_documents = state["documents"]

    if not filtered_documents:
        # 모든 문서가 관련성 없는 경우 질문 재작성
        print(
            "==== [DECISION: ALL DOCUMENTS ARE NOT RELEVANT TO QUESTION, TRANSFORM QUERY] ===="
        )
        return "transform_query"
    else:
        # 관련성 있는 문서가 있는 경우 답변 생성
        print("==== [DECISION: GENERATE] ====")
        return "generate"


def hallucination_check(state):
    """생성된 답변의 환각 여부와 질문 해결 여부를 순차적으로 평가하는 조건부 엣지 함수입니다.

    Args:
        state: question, documents, generation 필드를 포함한 그래프 상태

    Returns:
        다음 노드 이름 ("hallucination", "relevant", "not relevant")
    """
    print("==== [CHECK HALLUCINATIONS] ====")
    # 질문과 문서 검색 결과 가져오기
    question = state["question"]
    documents = state["documents"]
    generation = state["generation"]

    # 환각 평가
    score = hallucination_grader.invoke(
        {"documents": documents, "generation": generation}
    )
    grade = score.binary_score

    # Hallucination 여부 확인
    if grade == "yes":
        print("==== [DECISION: GENERATION IS GROUNDED IN DOCUMENTS] ====")

        # 답변의 관련성(Relevance) 평가
        print("==== [GRADE GENERATED ANSWER vs QUESTION] ====")
        score = answer_grader.invoke({"question": question, "generation": generation})
        grade = score.binary_score

        # 관련성 평가 결과에 따른 처리
        if grade == "yes":
            print("==== [DECISION: GENERATED ANSWER ADDRESSES QUESTION] ====")
            return "relevant"
        else:
            print("==== [DECISION: GENERATED ANSWER DOES NOT ADDRESS QUESTION] ====")
            return "not relevant"
    else:
        print("==== [DECISION: GENERATION IS NOT GROUNDED IN DOCUMENTS, RE-TRY] ====")
        return "hallucination"
```

### 그래프 컴파일

**그래프 컴파일** 단계에서는 **Adaptive RAG**의 워크플로우를 구축하고 실행 가능한 상태로 만듭니다. 이 과정은 그래프의 각 노드와 엣지를 연결하여 쿼리 처리의 전체 흐름을 정의합니다.

- **노드 정의**: 각 노드를 정의하여 그래프의 상태와 전환을 명확히 합니다.
- **엣지 설정**: 노드 간의 엣지를 설정하여 쿼리가 적절한 경로를 따라 진행되도록 합니다.
- **워크플로우 구축**: 그래프의 전체 흐름을 구축하여 정보 검색과 생성의 효율성을 극대화합니다.

```python
from langgraph.graph import END, StateGraph, START
from langgraph.checkpoint.memory import MemorySaver

# 그래프 상태 초기화
workflow = StateGraph(GraphState)

# 노드 정의
workflow.add_node("web_search", web_search)  # 웹 검색
workflow.add_node("retrieve", retrieve)  # 문서 검색
workflow.add_node("grade_documents", grade_documents)  # 문서 평가
workflow.add_node("generate", generate)  # 답변 생성
workflow.add_node("transform_query", transform_query)  # 쿼리 변환

# 그래프 빌드
workflow.add_conditional_edges(
    START,
    route_question,
    {
        "web_search": "web_search",  # 웹 검색으로 라우팅
        "vectorstore": "retrieve",  # 벡터스토어로 라우팅
    },
)
workflow.add_edge("web_search", "generate")  # 웹 검색 후 답변 생성
workflow.add_edge("retrieve", "grade_documents")  # 문서 검색 후 평가
workflow.add_conditional_edges(
    "grade_documents",
    decide_to_generate,
    {
        "transform_query": "transform_query",  # 쿼리 변환 필요
        "generate": "generate",  # 답변 생성 가능
    },
)
workflow.add_edge("transform_query", "retrieve")  # 쿼리 변환 후 문서 검색
workflow.add_conditional_edges(
    "generate",
    hallucination_check,
    {
        "hallucination": "generate",  # Hallucination 발생 시 재생성
        "relevant": END,  # 답변의 관련성 여부 통과
        "not relevant": "transform_query",  # 답변의 관련성 여부 통과 실패 시 쿼리 변환
    },
)

# 그래프 컴파일
app = workflow.compile(checkpointer=MemorySaver())
```

## 그래프 사용

**그래프 사용** 단계에서는 **Adaptive RAG**의 실행을 통해 쿼리 처리 결과를 확인합니다. 이 과정은 그래프의 각 노드와 엣지를 따라 쿼리를 처리하여 최종 결과를 생성합니다.

- **그래프 실행**: 정의된 그래프를 실행하여 쿼리의 흐름을 따라갑니다.
- **결과 확인**: 그래프 실행 후 생성된 결과를 검토하여 쿼리가 적절히 처리되었는지 확인합니다.

## 정리

이 튜토리얼에서는 **Adaptive RAG** 전략을 LangGraph를 활용하여 구현했습니다. 쿼리 분석을 통한 동적 라우팅과 자기 수정(Self-Reflective) RAG를 결합하여 다양한 유형의 질문에 효과적으로 대응하는 시스템을 구축했습니다.

### 핵심 구성 요소 요약

1. **쿼리 라우터(Question Router)**: 질문의 성격을 분석하여 웹 검색 또는 벡터스토어 검색으로 최적 경로를 선택합니다.
2. **검색 평가기(Retrieval Grader)**: 검색된 문서의 관련성을 평가하여 불필요한 문서를 필터링합니다.
3. **RAG 체인(RAG Chain)**: 검색된 문서를 컨텍스트로 활용하여 LLM이 정확한 답변을 생성합니다.
4. **Hallucination 체커**: 생성된 답변이 검색된 문서에 근거하는지 검증합니다.
5. **답변 관련성 평가기(Answer Grader)**: 답변이 사용자 질문을 실제로 해결하는지 최종 검증합니다.
6. **쿼리 재작성기(Query Rewriter)**: 검색 실패 시 질문을 재작성하여 검색 품질을 향상시킵니다.
7. **웹 검색(Web Search)**: 최신 정보가 필요한 경우 Tavily API를 통해 웹 검색을 수행합니다.
